## 백터화 연산
- 넘파이는 벡터화 연산을 지원한다.
- 벡터화 연산을 사용하면 명시적으로 반복문을 사용하지 않고도 배열의 모든 원소에 대해 반복 연산을 할 수 있다.
- 벡터화 연산은 선형 대수 공식과 동일한 간단한 파이썬 코드를 작성할 수 있다. 

예를 들어, 선형 대수에서 두 벡터의 합은 다음과 같이 구한다.




일 때, 두 벡터의 합 $z = x + y$ 은 다음과 같이 구한다.



만약, 벡터화 연산을 사용하지 않는다면 이 연산은 반복문을 사용하여 다음과 같이 만들어야 한다.

이 코드에서  `%time`은 셀 코드의 실행시간을 측정하는 Ipython magic 명령이다.

````python
import numpy as np
````

````python
x = np.arange(1,10001)
y = np.arange(10001,20001)
````

````python
%%time 
z = np.zeros_like(x)
for i in range(10000):
    z[i] = x[i]+y[i]
z[:10]
````

벡턱화 연산을 사용하면 덧셈 연산 하나로 끝난다. 

위에서 본 선형 대수의 벡터 기호를 사용한 연산과 결과가 동일하다.

연산 속도도 벡터화 연산이 훨씬 빠르다.

````python
%%time
z = x + y
z[:10]
````

- 사칙 연산 뿐 아니라, 비교 연산과 같은 논리 연산도 벡터화 연산이 가능하다.

````python
a = np.array([1,2,3,4])
b = np.array([4,2,2,4])
````

````python
a == b
````

````python
a >= b
````

- 만약 배열의 각 원소들을 일일히 비교하는 것이 아니라, 배열의 모든 원소가 다 같은지 알고 싶다면 `all` 명령을 사용하면 된다.

````python
a = np.array([1,2,3,4])
b = np.array([4,2,2,4])
c = np.array([1,2,3,4])
````

````python
np.all(a==b)
````

````python
np.all(a==c)
````

- 지수함수, 로그 함수 등의 수학 함수도 벡터화 연산을 지원한다. 

````python
a = np.arange(5)
a
````

````python
# np.exp() : 밑이 자연상수 e인 지수함수(e^x)로 변환
np.exp(a) # e^0, e^1, e^2, e^3, e^4
````

````python
10 ** a
````

````python
# np.log() : 요소 단위로 입력 어레이의 자연 로그 (natural logarithm) 값을 반환 (밑이 10인 log)
# log 밑이 2인 함수 == np.log2()
np.log(a+1)
````

## 브로드캐스팅 (broadcasting)
- 벡터/행렬 끼리 덧셈 혹은 뺼셈을 하려면 두 벡터/행렬의 크기가 같아야 한다. 
- 넘파이에서는 서로 다른 크기를 가진 두 배열의 사칙 연산도 지원함 ==> 브로드 캐스팅
- 크기가 작은 배열을 자동으로 반복 확장하여 크기가 큰 배열에 맞추는 방법

![img](https://dschloe.github.io/img/python/basic_syntax/broadcasting.png)

** 참조 ) https://dschloe.github.io/python/python_edu/01_basic/chapter_1_2_eda_with_numpy_broadcasting/

예를 들어 다음과 같이 벡터와 스칼라를 더하는 경우를 생각해보면,

브로드 캐스팅은 다음과 같이 스칼라를 벡터와 같은 크기로 확장시켜서 덧셈 계산을 하는 것!


````python
x = np.arange(5)
x
````

````python
y = np.ones_like(x)
y
````

````python
x + y
````

````python
x + 1
````

- 브로드캐스팅은 다음처럼 더 차원이 높은 경우에도 적용된다.



````python
x = np.vstack([range(7)[i:i +3] for i in range(5)])
x
````

````python
y = np.arange(5)[:,np.newaxis]
y
````

````python
x + y
````

````python
y = np.arange(3)
y
````

````python
x + y
````

## 차원 축소 연산
- 행렬의 하나의 행에 있는 원소들을 하나의 데이터 집합으로 보고 그 집합의 평균을 구하면 각 행에 대해 하나의 숫자가 나오게 된다. 
- 예를 들어 10x5 크기의 2차원 배열에 대해 행 평균을 구하면 10개의 숫자를 가진 1차원 벡터가 나오게 된다. 
- 이러한 연산을 차원 축소 (dimension reduction)연산이라고 한다.


- 넘파이는 다음과 같은 차원 축소 연산 명령 혹은 메서드를 지원한다.
    - 최대 / 최소 : min, max, argmin(최소값의 위치), argmax(최대값의 위치)
    - 통계 : sum, mean, median, std, var
    - 불리언 : all, any

````python
x = np.array([1,2,3,4])
x
````

````python
np.sum(x)
````

````python
x.sum()
````

````python
x = np.array([1,3,2])
x
````

````python
x.min()
````

````python
min(x)
````

````python
x.max()
````

````python
max(x)
````

````python
x.argmin() 
````

````python
x.argmax() 
````

````python
x = np.array([1,2,3,1])
x
````

````python
x.mean()
````

````python
np.median(x)
````

````python
np.all([True,True,False])
````

````python
np.any([True,True,False])
````

````python
a = np.zeros((100,100),dtype = np.int)
a
````

````python
np.any(a != 0)
````

````python
np.all(a == a)
````

````python
a = np.array([1,2,3,2])
b = np.array([2,2,3,2])
c = np.array([6,4,4,5])
````

````python
((a <= b) & (b <=c)).all()
````

- 연산의 대상이 2차원 이상인 경우에는 어느 차원으로 계산을 할지를 `axis` 인수를 사용하여 지시한다.
- `axis= 0`인 경우는 열 연산, `axis= 1`인 경우는 행 연산이다. 
- 디폴트 값은 `axis= 0`
- `axis` 인수는 대부분의 차원 축수 명령에 적용할 수 있다. 

````python
x = np.array([[1,1],[2,2]])
x
````

````python
x.sum()
````

````python
x.sum(axis = 0) # 열 합계
````

````python
x.sum(axis = 1) # 행 합계
````

실수로 이루어진 5x6형태의 데이터 행렬을 만들고 이 데이터에 대해 다음과 같은 값을 구한다. 
```
1. 전체의 최댓값
```

## 정렬
- `sort`함수나 메서드를 사용하여 배열 안의 원소를 크기에 따라 정렬하여 새로운 배열을 만들 수 있다. 
- 2차원 이상인 경우에는 행이나 열을 각각 따로따로 정렬하는데 `axis`인수를 사용하여 행을 정렬할 것인지 열을 정렬할 것인지 결정한다. 
- `axis=0`이면 각각의 행을 따로따로 정렬하고 `axis=1`이면 각각의 열을 따로따로 정렬한다. 
- 디폴트 값은 -1, 가장 안쪽(나중)의 차원

````python
a = np.array([[4,3,5,7],
              [1,12,11,9],
              [2,15,1,14]])
a
````

````python
np.sort(a) # axis = -1 또는 axis = 1과 동일 
````

````python
np.sort(a,axis = 0)
````

````python
a
````

- `sort`메서드는 해당 객체의 자료 자체가 변화하는 자체변화 (in-place)메서드이므로 사용할 때 주의를 기울여야 한다. 

````python
a.sort(axis = 1)
a
````

- 만약 자료를 정렬하는 것이 아니라 순서만 알고 싶다면 `argsort`명령을 사용한다.

````python
a = np.array([42,38,12,25])
j = np.argsort(a)
j
````

````python
a[j]
````

````python
np.sort(a)
````

````python
a
````

````python
a.sort()
a
````

### 연습문제
다음 배열은 첫번째 행(row)에 학번, 두번째 행에 영어 성적, 세번째 행에 수학 성적을 적은 배열이다.<br>
영어성적을 기준으로 각 열(column)을 재정렬하라.
```
array([[ 1,  2,   3,   4],
       [46, 99, 100,  71],
       [81, 59,  90, 100]])
```

````python
score = np.array([[ 1,  2,   3,   4],
       [46, 99, 100,  71],
       [81, 59,  90, 100]])
````

````python
score = score.T
eng_sort_idx = np.argsort(score[:,1])
score[eng_sort_idx].T
````
