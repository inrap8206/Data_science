> Ïù¥Îü¨Ìïú Í∞íÎì§ÏùÑ ÌÜµÌãÄÏñ¥ Í∏∞Ïà† ÌÜµÍ≥Ñ(descriptive statistics)ÎùºÍ≥† ÌïúÎã§.

xÎ•º Ïù¥Î£®Îäî Ïà´Ïûê ÌïòÎÇòÌïòÎÇòÎ•º ÏàòÌïô Í∏∞Ìò∏Î°úÎäî $ùë•1,ùë•2,‚ãØ,ùë•ùëÅ$ Ï≤òÎüº ÌëúÏãúÌïúÎã§. ÏúÑ ÏòàÏóêÏÑú $x1=18,  x2=15$

```python
import numpy as np
```

```python
x = np.array([18,   5,  10,  23,  19,  -8,  10,   0,   0,   5,   2,  15,   8,
              2,   5,   4,  15,  -1,   4,  -7, -24,   7,   9,  -6,  23, -13])
```

ÌèâÍ∑†ÏùÑ ÌÜµÍ≥ÑÏö©Ïñ¥Î°úÎäî ÌëúÎ≥∏ ÌèâÍ∑†(sample average, sample mean)Ïù¥ÎùºÍ≥† ÌïúÎã§.<br>
ùë• Îç∞Ïù¥ÌÑ∞Ïóê ÎåÄÌïú ÌëúÎ≥∏ ÌèâÍ∑†ÏùÄ ùë•¬Ø ÎùºÍ≥† ÌëúÏãúÌïòÎ©∞ Îã§ÏùåÍ≥º Í∞ôÏù¥ Í≥ÑÏÇ∞ÌïúÎã§.<br> 
Ïù¥ ÏãùÏóêÏÑú ùëÅÏùÄ Îç∞Ïù¥ÌÑ∞Ïùò Í∞úÏàòÏù¥Îã§.



```python
np.mean(x)
```

### Î∂ÑÏÇ∞ (variance)

ÌëúÎ≥∏ Î∂ÑÏÇ∞(sample variance)ÏùÄ Îç∞Ïù¥ÌÑ∞ÏôÄ ÌëúÎ≥∏ ÌèâÍ∑†Í∞ÑÏùò Í±∞Î¶¨Ïùò Ï†úÍ≥±Ïùò ÌèâÍ∑†Ïù¥Îã§.<br> 
ÌëúÎ≥∏ Î∂ÑÏÇ∞Ïù¥ ÏûëÏúºÎ©¥ Îç∞Ïù¥ÌÑ∞Í∞Ä Î™®Ïó¨ÏûàÎäî Í≤ÉÏù¥Í≥† ÌÅ¨Î©¥ Ìù©Ïñ¥Ï†∏ ÏûàÎäî Í≤ÉÏù¥Îã§.<br>
ÏàòÌïô Í∏∞Ìò∏Î°úÎäî ùë†2Ïù¥ÎùºÍ≥† ÌëúÏãúÌïòÎ©∞ Îã§ÏùåÍ≥º Í∞ôÏù¥ Í≥ÑÏÇ∞ÌïúÎã§.


```python
np.var(x)
```

```python
np.var(x, ddof=1)  # ÎπÑÌé∏Ìñ• Î∂ÑÏÇ∞
```

ÌëúÎ≥∏ ÌëúÏ§ÄÌé∏Ï∞®(sample standard variance)Îäî ÌëúÎ≥∏ Î∂ÑÏÇ∞Ïùò ÏñëÏùò Ï†úÍ≥±Í∑º Í∞íÏù¥Îã§.<br>
ùë†Ïù¥ÎùºÍ≥† ÌëúÏãúÌïúÎã§.

### ÏÇ¨Î∂ÑÏúÑÏàò(quartile)

ÏÇ¨Î∂ÑÏúÑÏàò(quartile)Îäî Îç∞Ïù¥ÌÑ∞Î•º Í∞ÄÏû• ÏûëÏùÄ ÏàòÎ∂ÄÌÑ∞ Í∞ÄÏû• ÌÅ∞ ÏàòÍπåÏßÄ ÌÅ¨Í∏∞Í∞Ä Ïª§ÏßÄÎäî ÏàúÏÑúÎåÄÎ°ú Ï†ïÎ†¨ÌïòÏòÄÏùÑ Îïå 1/4, 2/4, 3/4 ÏúÑÏπòÏóê ÏûàÎäî ÏàòÎ•º ÎßêÌïúÎã§.<br>
Í∞ÅÍ∞Å 1ÏÇ¨Î∂ÑÏúÑÏàò, 2ÏÇ¨Î∂ÑÏúÑÏàò, 3ÏÇ¨Î∂ÑÏúÑÏàòÎùºÍ≥† ÌïúÎã§.<br>
1/4Ïùò ÏúÑÏπòÎûÄ Ï†ÑÏ≤¥ Îç∞Ïù¥ÌÑ∞Ïùò ÏàòÍ∞Ä ÎßåÏïΩ 100Í∞úÏù¥Î©¥ 25Î≤àÏß∏ ÏàúÏÑú, Ï¶â ÌïòÏúÑ 25%Î•º ÎßêÌïúÎã§.<br>
Îî∞ÎùºÏÑú 2ÏÇ¨Î∂ÑÏúÑÏàòÎäî Ï§ëÏïôÍ∞íÍ≥º Í∞ôÎã§.

ÎïåÎ°úÎäî ÏúÑÏπòÎ•º 1/100 Îã®ÏúÑÎ°ú ÎÇòÎàà Î∞±Î∂ÑÏúÑÏàò(percentile)ÏùÑ ÏÇ¨Ïö©ÌïòÍ∏∞ÎèÑ ÌïúÎã§. 1ÏÇ¨Î∂ÑÏúÑÏàòÎäî 25% Î∞±Î∂ÑÏúÑÏàòÏôÄ Í∞ôÎã§.
